```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
#cv2.data.haarcascades is a folder path where OpenCV stores it's pre-trained detector files
#this is a pre-trained face detector that comes built into OpenCV
#.xml is a file that contains many rules OpenCV already learned from analyzing millions of faces

eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
#also a pre-built detector for eyes

def show_image(img_bgr, title = ""):  #reusable function to call later, img_bgr is the image that you pass in
  img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
  #this converts BGR(Blue, Grreen, Red) to RGB order becasue OpenCV loads images in BGR but matplotlib requires RGB

  plt.figure(figsize = (6,6))
  plt.imshow(img_rgb)
  plt.title(title)
  plt.axis("off")
  plt.show()
```

```python
img = cv2.imread("Barack_Obama.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5)

for (x, y, w, h) in faces:
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

show_image(img, "Barack Obama - Face Detection")
```
