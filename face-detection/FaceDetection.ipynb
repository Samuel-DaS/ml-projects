```python
import numpy as np  #library for handling numbers and arrays
import cv2  #library for computer vision, such as loading
import matplotlib.pyplot as plt #for plotting and visualization
%matplotlib inline
#the "%" means this is a special colab/Jupyter command,
#inline will allow images and charts to open directly in the notebook
```

```python
Obama1 = cv2.imread("Barack_Obama.jpg", 0)
Clinton = cv2.imread("Bill_Clinton.jpg", 0)
Bush = cv2.imread("George_W_Bush.jpg", 0)
Former_Presidents = cv2.imread("Former_Presidents.jpg", 0)

#imread is a function from the cv2 library for image read
#adding the 0 at the end will load the image in grayscale instead of color
#no color adds less data, which is better for Haar Cascade to run faster
```

```python
Obama = cv2.imread("Barack_Obama.jpg", cv2.IMREAD_GRAYSCALE) #cv2.IMREAD_GRAYSCALE will load image in grayscale

if Obama is None:               #if image wasnt found, Obama would be None
  print("Failed to load image")
else:
  plt.figure(figsize = (6,6)) #creats a 6,6 display
  plt.imshow(Obama, cmap = "gray")  #tells matplotlib to render it as grayscale image, or else would be a blue tint
  plt.title("Barack Obama - Grayscale")
  plt.axis("off")
  plt.show()
```

```python
Clinton = cv2.imread("Bill_Clinton.jpg", cv2.IMREAD_GRAYSCALE)

if Clinton is None:
  print("Failed to load image.")
else:
  plt.figure(figsize = (6,6))
  plt.imshow(Clinton, cmap = "gray")
  plt.title("Bill Clinton - Grayscale")
  plt.axis("off")
  plt.show()
```
